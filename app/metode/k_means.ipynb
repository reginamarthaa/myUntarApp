```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.metrics import silhouette_score

from sklearn.model_selection import train_test_split
from sklearn.svm import OneClassSVM
```

```python
# memanggil dataset

dataset = pd.read_csv("sample_data/data_latih_kuesioner.csv", index_col=None)
```

```python
# menampilkan data

data_latih = dataset
data_latih
```

```python
X = data_latih

# membuat function k-means untuk mencari hasil terbaik

def k_means_clustering(random_states):

    kmeans = KMeans(n_clusters=3, random_state=random_states)
    labels = kmeans.fit_predict(X)

    Y = labels

    silhouette_avg = silhouette_score(X, Y)
    return X, Y, silhouette_avg
```

```python
# membuat looping untuk mencari hasil terbaik

temp = 0
index = 0

for i in np.arange(0,100):
    X, Y, silhouette_avg = k_means_clustering(i)
    if silhouette_avg > temp:
        temp = silhouette_avg
        index = i
        print(f"temp: {temp}")
        print (f"index: {index}")
```

```python
# dari hasil looping, diambil nilai Y menggunakan random state terbaik untuk menentukan label dari dataset

X, Y, silhouette_avg = k_means_clustering(index)

print(silhouette_avg)
```

```python
# menggabungkan fitur dengan label dari dataset

data_berlabel = data_latih
data_berlabel["label"] = Y
print(data_berlabel)
```

```python
# menyimpan hasil dataset yang sudah diberi label
data_berlabel.to_csv(f'dataset_rs{index}_{data_berlabel.shape}.csv', index=False)
```
